```python
from sklearn.model_selection import learning_curve
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# 加载参数
digits = load_digits()
X = digits.data
y = digits.target
#gramma  = 0.001
train_size , train_scores,val_scores = learning_curve(
    SVC(gamma=0.001),X,y,cv=10,scoring='accuracy',
    train_sizes=[0.1,0.25,0.5,0.75,1])
# 在十折的交叉验证数据上进行平均
train_scores_mean = np.mean(train_scores,axis=1)
val_scores_mean = np.mean(val_scores,axis=1)
# 绘制学习曲线
plt.plot(train_size,train_scores_mean,'o-',color='r',label='training')
plt.plot(train_size,val_scores_mean,'*-',color='g',label='cross validation')

plt.xlabel('training sample size')
plt.ylabel('accuracy')
plt.legend(loc='best')
plt.show()
```

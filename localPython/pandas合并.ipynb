```python
import pandas as pd
import numpy as np
```

```python
dfl = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['a','b','c','d'])
df2 = pd.DataFrame(np.arange(12,24).reshape((3,4)),columns=['a','b','c','d'])
df3 = pd.DataFrame(np.arange(24,36).reshape((3,4)),columns=['a','b','c','d'])
print(dfl)
print(df2)
print(df3)
```

```python
df4 = pd.concat([dfl,df2,df3],axis=0,ignore_index=True)#纵向合并,不考虑原来的index
```

```python
dfl = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['a','b','c','f'])
df2 = pd.DataFrame(np.arange(12,24).reshape((3,4)),columns=['a','c','d','e'])
print(dfl)
print(df2)
```

```python
df6 = pd.concat([dfl,df2],join='outer',ignore_index=True)#合并两个表，缺少的部分填充NaN
df6
```

```python
df7 = pd.concat([dfl,df2],join='inner',ignore_index=True)
df7
```

```python
dfl = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['a','b','c','f'])
df2 = pd.DataFrame(np.arange(12,24).reshape((4,3)),columns=['a','c','d'])
print(dfl)
print(df2)
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
print(left)
print(right)
```

```python
res = pd.merge(left,right,on='key')
res
```

```python
left = pd.DataFrame({'key1':['K0','K0','K2','K3'],
                     'key2':['K0','K1','K0','K0'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K1','K3'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
print(left)
print(right)
```

```python
# how =['left','right','inner','outer']
res = pd.merge(left,right,on=['key1','key2'],how='outer')#默认inner
res
```

```python
# how =['left','right','inner','outer']
res = pd.merge(left,right,on=['key1','key2'],how='inner')#默认inner 只合并key1,key2都存在的行，不同就舍弃
res
```

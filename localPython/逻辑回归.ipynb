```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-60,60,200)
y = [1/(1+math.e**(-x)) for x in X ]
plt.plot(X,y)
plt.show()
```

```python
X = np.linspace(0.0001,1,200)
y = [(-np.log(x)) for x in X]
plt.plot(X,y)
plt.show()
```

```python
X = np.linspace(0.9999,1,200)
y = [(-np.log(x)) for x in X]
plt.plot(X,y)
plt.show()
```
